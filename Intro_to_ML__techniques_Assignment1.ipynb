```python
# Gradient Descent function for x,y,aplha(learning rate) and epochs(no.of iterations) and value os a and b for y=a^x+b

def gradient_descent(x,y):
    a_i = 0    # initial value of a
    b_i = 0    # initial value of b
    epochs = 1000000  # number of iterations
    n = len(x)
    alpha = 0.000001   # learning rate

    cost_previous = 0

    for i in range(epochs):
        y_pred = a_i * x + b_i
        cost = (1/n)*sum([value**2 for value in (y-y_pred)])   # Mean square error
        a_dif = -(2/n)*sum(x*(y-y_pred))
        b_dif = -(2/n)*sum(y-y_pred)
        a_i = a_i - alpha * a_dif
        b_i = b_i - alpha * b_dif
        cost_previous = cost
        print ("a {}, b {}, cost {}, epochs {}".format(a_i,b_i,cost, i))

    return a_i, b_i

if __name__ == "__main__":
    uploaded=files.upload()
    df = pd.read_csv("data.csv")
    x = np.array(df.x)
    y = np.array(df.y)

    a_i, b_i = gradient_descent(x,y)
    print("Using gradient descent function: a {} , b {}".format(a_i, b_i))
```

```python
# Plot the histogram with the residuals (y- ŷ) after model training for the original and predicted values

y_pred = a_i * x + b_i
residuals = y - y_pred
plt.hist(residuals,edgecolor='black',label='Residuals' ,bins=30)
plt.legend()
```

```python
# prompt: Plot the ŷ vs x (line chart), preferrably on top of a y vs x scatter plot

plt.scatter(df.x,df.y,color='red')
plt.plot(x, y_pred, color='blue')
plt.xlabel('x')
plt.ylabel('y')
```
